##Tratamiento de Missing

#### Verificar la presencia de valores nulos

```python
nulos = df2.isnull().sum()
```

###Imputar o eliminar valores nulos según sea necesario
###Ejemplo de imputación media para columnas numéricas
###Calcular la media solo para columnas numéricas

```python
mean_values = df2.mean(numeric_only=True)
```

###Llenar valores nulos con las medias respectivas


```python
df2.fillna(mean_values, inplace=True)
```

```python
df2.dropna(inplace=True)
print(nulos)
```

##Identificación y Eliminación de Errores

###Identificar posibles errores en columnas específicas
###Supongamos que 'votos' no puede ser un número negativo
###Iterar sobre todas las columnas y verificar posibles errores

```python
for col in df2.columns:
    if df2[col].dtype == 'float64':  # Asegurarse de que sea una columna numérica
        errores = df2[df2[col] < 0]
        if not errores.empty:
            print(f"Errores en la columna {col}:")
            print(errores)
```
